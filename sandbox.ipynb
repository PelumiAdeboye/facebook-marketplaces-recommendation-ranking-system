```python
import pandas as pd
```

```python
# Step 1: Read the "Products.csv" and "Images.csv" files
products_df = pd.read_csv("cleaned_tabular_data.csv")
images_df = pd.read_csv("Images.csv")
```

```python
images_df.head(20)
```

```python
# Step 2: Create a function to extract the root category from the original category
def extract_root_category(category):
    # Assuming the categories are separated by "/"
    categories = category.split("/")
    return categories[0].strip()
```

```python
# Step 3: Apply the extract_root_category function to create a new column "root_category" in the products_df
products_df["root_category"] = products_df["category"].apply(extract_root_category)
```

```python
product_df = products_df.drop(products_df.columns[0],axis=1)
product_df.head(10)
```

```python
# Step 4: Create a list of unique root categories and their corresponding labels
unique_root_categories = product_df["root_category"].unique()
unique_root_categories
```

```python
category_labels = {category: label for label, category in enumerate(unique_root_categories)}
```

```python
# Step 5: Print the unique root categories and their corresponding labels
print(category_labels)
```

```python
# Step 3: Merge the dataframes to obtain the appropriate category for each image ID
merged_df = pd.merge(images_df, product_df, left_on="product_id", right_on="id", how="inner")
```

```python
# Step 4: Create a new column "labels" in the "Images.csv" dataframe
#merged_df["labels"] = merged_df["category"].map(category_labels)
```

```python
merged_df["labels"]= merged_df["root_category"].replace(category_labels)
```

```python
merged_df.head(20)
```

```python
merged_df.info()
```

```python
# Step 5: Save the updated "Images.csv" dataframe as "training_data.csv"
merged_df.to_csv("training_data.csv", index=False)
```
